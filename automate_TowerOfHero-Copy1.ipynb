```python
def open_excel():
    pgui.click(x=1862, y=53,duration=1)
    time.sleep(0.1)
    pgui.write("excel",0.3)
    time.sleep(0.8)
    pgui.hotkey("return")
```

```python
open_excel()
pgui.hotkey("option", "up") # 一番上のファイルを開く
pgui.hotkey("enter")
```

```python
def btm2top():
    # 下に移動
    # x=1138, y=973
    pyautogui.moveTo(1138,973)

    # 上にドラッグ
    # x=1138, y=720
    pyautogui.dragTo(1138, 720, 0.2)
```

```python
btm2top()
```

```python
def top2btm():
    # 上に移動
    # x=1138, y=720
    pyautogui.moveTo(1138, 720)
    
    # 下にドラッグ
    # x=1138, y=973
    pyautogui.dragTo(1138,973, 0.2)

    
    
```
